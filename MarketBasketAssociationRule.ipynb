```python
# creating a function to remove extra whitespace from data
import re

for i in df.columns:
    if df[i].dtype == 'object':
        df[i] = df[i].astype('str').apply(lambda x: re.sub(r'\s+', ' ', x).strip())
    else:
        pass

```

```python
df.head(5)
```

```python
# convert dataframe from wide format to long format containing Basket and Product Description
df1 = pd.melt(df, id_vars = 'Basket', value_name = 'Product Description')
df1.drop(columns = 'variable', inplace = True)
df1
```

```python
# pivot dataframe to convert each product item into columns
grouped_df1 = df1.groupby(['Basket', 'Product Description']).size().unstack().fillna(0).reset_index().set_index('Basket')
grouped_df1
```

```python
def custom_encoding(x):
    if x <= 0:
        return 0
    if x > 0:
        return 1

basket_input = grouped_df1.applymap(custom_encoding)
basket_input.drop('nan', inplace = True, axis = 1)
```

```python
frequent_itemsets = apriori(basket_input, min_support = 0.05, use_colnames = True)
```

```python
rules=association_rules(frequent_itemsets, metric = 'lift')
```

```python
rules.sort_values(by = 'lift', ascending = False)
```
